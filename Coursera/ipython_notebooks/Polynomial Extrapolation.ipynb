```python
import numpy as np
```

```python
%matplotlib inline
from matplotlib import pylab as plt
```

```python
x = np.arange(1, 15, 0.01)
y = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
plt.plot(x,y)
plt.show()
```

```python
from scipy import linalg

def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

x1 = np.array([[1, 1], [1, 15]])
func = np.array([f(1), f(15)])
w1 = linalg.solve(x1,func)

pol1 = w1[0] + w1[1] * x
plt.plot(x,pol1)
plt.show()
```

```python
x2 = np.array([[1, 1, 1], [1, 8, 8**2], [1, 15, 15**2]])
print(x2)
func = np.array([f(1), f(8), f(15)])
w2 = linalg.solve(x2,func)
print(w2)
pol2 = w2[0] + w2[1] * x + w2[2] * x ** 2
plt.plot(x,pol2)
plt.show()
```

```python
x3 = np.array([[1, 1, 1, 1], [1, 4, 4**2, 4**3], [1, 10, 10**2, 10**3], [1, 15, 15**2, 15**3]])
func = np.array([f(1), f(4), f(10), f(15)])
w3 = linalg.solve(x3,func)

pol3 = w3[0] + w3[1] * x + w3[2] * x ** 2 + w3[3] * x ** 3
plt.plot(x,y)
plt.plot(x,pol1)
plt.plot(x,pol2)
plt.plot(x,pol3)
plt.show()
print(w3)
```
